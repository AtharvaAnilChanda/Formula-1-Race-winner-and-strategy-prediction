**Reading the CSV**

```python
data = pd.read_csv("../input/f1-withgrid/WithGrid55.csv")
train = pd.read_csv("../input/traindata/Train.csv")
test = pd.read_csv("../input/testdata1/Test.csv")
#test = pd.read_csv("../input/scamdata/scam.csv")
```

**Adding Prediction Column**

```python
train['Class'] = ''
for index, row in train.iterrows():
    if(row['GridNo'] < 4):
        train['Class'][index] = 0
    elif(4 <= row['GridNo'] < 10):
        train['Class'][index] = 1
    else:
        train['Class'][index] = 2
#test = test.sample(frac = 1, random_state = 1).reset_index()
#train = train.sample(frac = 1, random_state = 1).reset_index()
train
```

**Encoding Categorical Data**

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
#sc  = StandardScaler()
le = LabelEncoder()
#X_Train=train[['Session','Driver','Team','GridNo']]
X_Train=train[['Session','Driver','Team']]
#X_Train=train[['Driver','Team']]
X_Train['Session'] = le.fit_transform(X_Train['Session'])
X_Train['Driver'] = le.fit_transform(X_Train['Driver'])
X_Train['Team'] = le.fit_transform(X_Train['Team'])
X_Train
```

```python
#X_Test = test[['Session','Driver','Team','GridNo']]
X_Test = test[['Session','Driver','Team']]
#X_Test = test[['Driver','Team']]
X_Test['Session'] = le.fit_transform(X_Test['Session'])
X_Test['Driver'] = le.fit_transform(X_Test['Driver'])
X_Test['Team'] = le.fit_transform(X_Test['Team'])
X_Test
```

```python
Y_Train = train[['Class']]
Y_Train = Y_Train.astype('int')
Y_Train
```

**Creating a Model**

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 100)
model.fit(X_Train, Y_Train)
Y_test = model.predict(X_Test)
for i in range(0, len(Y_test)):
    print(test['Driver'][i], Y_test[i])
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(X_Train, Y_Train)
Y_test = model.predict(X_Test)
for i in range(0, len(Y_test)):
    print(test['Driver'][i], Y_test[i])
```

```python
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(X_Train, Y_Train)
Y_test = model.predict(X_Test)
for i in range(0, len(Y_test)):
    print(test['Driver'][i], Y_test[i])
```
