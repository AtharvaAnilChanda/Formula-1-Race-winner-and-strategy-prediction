**Reading The Updated CSV**

```python
Train = pd.read_csv("../input/traindataset/ModifiedRaceStrategy.csv")
Train
```

**Getting the Explanatory Variables**

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
le = LabelEncoder()
X_Train = Train[['DriverNumber','LapNumber','Compound', 'FreshTyre', 'Stint', 'Team', 'Driver']]
X_Train['Compound'] = le.fit_transform(X_Train['Compound'])
X_Train['FreshTyre'] = le.fit_transform(X_Train['FreshTyre'])
X_Train['Driver'] = le.fit_transform(X_Train['Driver'])
X_Train['Team'] = le.fit_transform(X_Train['Team'])
X_Train
```

```python
Y_Train = Train[['TyreLife']]
Y_Train
```

**Test Data**

```python
Test = pd.read_csv("../input/teststrategy/TestDataset.csv")

from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
le = LabelEncoder()
X_Test = Test[['DriverNumber','LapNumber', 'Compound', 'FreshTyre', 'Stint', 'Team', 'Driver']]
X_Test['Compound'] = le.fit_transform(X_Test['Compound'])
X_Test['FreshTyre'] = le.fit_transform(X_Test['FreshTyre'])
X_Test['Driver'] = le.fit_transform(X_Test['Driver'])
X_Test['Team'] = le.fit_transform(X_Test['Team'])
X_Test = X_Test.dropna().reset_index(drop=True)
X_Test
```

**Predicting Model**

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators = 100, random_state = 0)
model.fit(X_Train, Y_Train)
Y_test = model.predict(X_Test)
soft = 0
softcount = 0
medium = 0
mediumcount = 0
hard = 0
hardcount = 0
print("Driver Tyre Actual Predicted")
for i in range(0, len(Test['Driver'])):
    if(Test['Compound'][i] == 'SOFT'):
        soft = soft + Y_test[i]
        softcount = softcount + 1
    elif(Test['Compound'][i] == 'MEDIUM'):
        medium = medium + Y_test[i]
        mediumcount = mediumcount + 1
    elif(Test['Compound'][i] == 'HARD'):
        hard = hard + Y_test[i]
        hardcount = hardcount + 1
    print(Test['Driver'][i], Test['Compound'][i], Test['TyreLife'][i], Y_test[i])
```

**Average of Each Tyre**

```python
softlife = soft/softcount
mediumlife = medium/mediumcount
hardlife = hard/hardcount
print("Soft:", softlife)
print("Medium:", mediumlife)
print("Hard:", hardlife)
print("Race 5 has 57 laps")
```
